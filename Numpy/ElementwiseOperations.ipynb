# Array Shape Manipilation
Flattening

```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
print(a.ravel()) #Returns a contiguous flattened array. A 1-D Array.
```

```python
print(a.T) # Transpose
```

```python
print(a.T.ravel())
```

Reshaping

```python
# The inverse operation to flattening
print(a.shape)
print(a)
```

```python
b = a.ravel()
print(b)
```

```python
b = b.reshape((2,3))
print(b.shape)
print(b)

b[1,1] = 10
print(b)
# Here value of a is also changing when we change b. So they are pointing to the same memory location.
print(a)
```

```python
a = np.arange(4)
a.resize((8,))
print(a)
```

Sorting Data

```python
# Sorting along an axis
a = np.array([[5,4,6],[2,3,2]])
b = np.sort(a, axis=1)
print(b)
```

```python
# In-place sort
a.sort(axis=1)
print(a)
```

```python
# Sorting with fancy indexing
a = np.array([4,3,1,2])
j = np.argsort(a) # Give index
print(j)
```

```python
print(a[j])
```
