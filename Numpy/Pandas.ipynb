Optimize Memory Usage for student Performance Grades

```python
print(f"Original 'grade' dtype: {df['grade'].dtype}")
print("Before optimization memory usage:")
print(df['grade'].memory_usage(deep=True))
df['grade'] = df['grade'].astype('category')
print(f"Optimized 'grade' dtype: {df['grade'].dtype}")
print("After optimization memory usage:")
print(df['grade'].memory_usage(deep=True))
```

5. Optimiize Storage for math score s Using Different Integer Types

```python
print(f"Original 'mathscore' dtype: {df['math_original'].dtype}")
print("Before optimization memory usage:")
print(df['math_original'].memory_usage(deep=True))

# convert math_original to int8 (correct conversion)
df['math_int8_fixed'] = df['math_original'].astype(np.int8)

# show corrected int8 dtype and memory usage
print(f"Optimized 'math_int8_fixed' dtype: {df['math_int8_fixed'].dtype}")
print("After optimization memory usage:")
print(df['math_int8_fixed'].memory_usage(deep=True))

# covert attendance  -> unit8(correct)
df['attendance_uint8'] = df['attendance'].astype(np.uint8)
print(f"'attendance_unit8' dtype: {df['attendance_unit8'].dtype}")
print("After optimization memory usage:")
print(df['attendance_uint8'].memory_usage(deep=True))
```

6. Explore Float Precision with science_original scores

```python
# 6. Explore Float Precision with science_original scores
print(f"Original 'science_scores' dtype: {df['science_original'].dtype}")
print("Before optimization memory usage:")
print(df['science_original'].memory_usage(deep=True))

# covert science_original  -> float16(correct)  for ML models we always use float16
df['science_scores'] = df['science_original'].astype(np.float16)
print(f"'science_scores' dtype: {df['science_scores'].dtype}")
print("After optimization memory usage:")
print(df['science_scores'].memory_usage(deep=True))

```

7. Optimize Memory for attendance Data

```python
print(f"Original 'attendance_Data' dtype: {df['attendance'].dtype}")
print("Before optimization memory usage:")
print(df['attendance'].memory_usage(deep=True))

# covert attendance  -> uint8(correct)
df['attendance_Data'] = df['attendance'].astype(np.uint8)
print(f"'attendance_Data' dtype: {df['attendance_Data'].dtype}")
print("After optimization memory usage:")
print(df['attendance_Data'].memory_usage(deep=True))
```

8. Compare Math Scores Between Different Student Groups

```python
math_scores_by_grade = df.groupby('grade')['math_original'].mean()
print("Average Math Scores By Grade:")
print(math_scores_by_grade)
```

Calculate Average Scores with Custom Weights

```python
math_weight = 0.4
science_weight = 0.4
english_weight = 0.4
df['custom_weighted_score'] = (df['math_original']*math_weight + df['science_original']*science_weight + df['english_original']*english_weight)
print("5 Custom weighted scores: ")
print(df[['math_original', 'science_original', 'english_original','custom_weighted_score']].head())
```
