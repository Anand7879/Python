```python
import pandas as pd
df = pd.read_csv("weather_data.csv")
df
```

```python
weather_data = [('1/1/2017',32,6,'Rain'),
                ('1/2/2017',35,6,'Sunny'),
                ('1/3/2017',28,6,'Snow'),
                ('1/4/2017',30,6,'Rain'),
                ('1/5/2017',31,6,'Snow'),
                ('1/6/2017',36,6,'Sunny')
                ]

df = pd.DataFrame(weather_data,columns=['day','temperature','windspeed','event'])
df
```
