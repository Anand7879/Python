# Elementwise Operations



1. Basic Operation

with scalars

```python
a = np.array([1,2,3,4]) 
a+1
```

```python
a**2
```

All arithmetic operates elementwise

```python
b = np.ones(4) +1
print(a)
print(b)

a-b
```

```python
a * b
```

```python
# matrix multiplication

c = np.diag([1,2,3,4])

print(c*c)
print("********************")
print(c.dot(c))
```

comparisions

```python
a = np.array([1,2,3,4])
b = np.array([5,2,3,6])

a==b
```

```python
a>b
```

```python
# array-wise comparisions
a = np.array([1,2,3,4])
b = np.array([5,2,3,7])
c = np.array([1,2,3,4])

np.array_equal(a,b)
```

```python
np.array_equal(a,c)
```

Logical Operations

```python
a = np.array([1,1,0,1],dtype=bool)
b = np.array([1,0,1,1],dtype=bool)

print(np.logical_or(a,b))
print(np.logical_and(a,b))
print(np.logical_not(a,b))
print(np.logical_xor(a,b))

```
