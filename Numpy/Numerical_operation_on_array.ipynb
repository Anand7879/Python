All arithmetic operates elementwise

```python
b = np.ones(4) +1
print(a)
print(b)

a-b
```

```python
# matrix multiplication

c = np.diag([1,2,3,4])

print(c*c)
print("********************")
print(c.dot(c))
```

comparisions

```python
a = np.array([1,2,3,4])
b = np.array([5,2,3,6])

a==b
```

```python
a>b
```

```python
# array-wise comparisions
a = np.array([1,2,3,4])
b = np.array([5,2,3,7])
c = np.array([1,2,3,4])

np.array_equal(a,b)
```

```python
np.array_equal(a,c)
```

Logical Operations

```python
a = np.array([1,1,0,1],dtype=bool)
b = np.array([1,0,1,1],dtype=bool)

print(np.logical_or(a,b))
print(np.logical_and(a,b))
print(np.logical_not(a,b))
print(np.logical_xor(a,b))

```

```python
a = np.array([1,2,0,1],dtype=bool)
a
```

```python
a = np.arange(5)
print(a)
print(np.sin(a))
print(np.cos(a))
print(np.tan(a))
print(np.log(a))
```

```python
np.exp(a) #evaluate e^x for each element in a given input
```

```python
# Computing Sum

x = np.array([1,2,3,4])
print(np.sum(x))
```

```python
# Sum by Rows and by columns

x = np.array([[1,1],[2,2]])
x
```

```python
x.sum(axis=0)  # columns first dimension
```

```python
x.sum(axis=1)  # row ( second dimension)
```

```python
a = np.array([1,2,3,4])
b = np.array([2,2,3,2])
c = np.array([6,4,4,5])
print(a<=b)
print(b<=c)
print((a<=b) & (b<=c))
print(((a<=b) & (b<=c)).all())
```
