# Numpy 

```python
import numpy as np
a = np.array([0,1,2,3,4])
print(a)

print(np.arange(10))
print(np.arange(5,10))
```

# why it is useful:
Memory-efficient container that provides fast numerical operations

```python
#python list
L = range(1000)
%timeit [i**2 for i in L]
```

```python
#with numpy
a = np.arange(1000)
%timeit a**2
```
