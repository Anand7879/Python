TUPLE Creation

```python
#empty tuple
t = ()

#tuple having integers
t = (1,2,3)
print(t)

#tuple with mixed datatypes
t = (1,'raju',28,'abc')
print(t)

#nested tuple
t = (1,(2,3,4),[1,'raju',28,'abc'])
print(*t)
```

```python
#only paranthesis is not enough
t = ('Anand')
type(t)
```

```python
#Need a comma at the end
t = ('Anand',)
type(t)
```

```python
#parenthesis is optional
t = "Anand",
print(type(t))
print(t)
```

Accessing Elements in Tuple

```python
t = ('Anand','Ankit','Divya','naveen')
print(t[1])
```

```python
#negative index
print(t[-1])  #print last element
```

```python
#nested tuple
t = ('Abc',('Anand','Ankit','Divya','naveen'))
print(t[1])
```

```python
print(t[1][2])
```

```python
#slicing 
t = (1,2,3,4,5,6)
print(t[1:4])

#print elements from starting to 2nd last elements
print(t[:-2])

#print elements from starting to end
print(t[:])
```

```python
#creating tuple
t = (1,2,3,4,[5,6,7])

# t[2] = 'x'
```

```python
#concatination

t1 = (1,2,3)
t2 = (5,6,7)
t3 = t1+t2
print(t3)
```

```python
#repeat the elements in a tuple for a given number of times using the * operator
t = (('satish',)*4)
print(t)
```

Deletion In Tuple

```python
# we can not change the elements in a tuple
#that also means we cannot delete or remove items from  a tuple

#delete entire tuple using del keyword
t = (1,2,3,4,5,6)

#delete entire tuple
del t
```

Tuple Count

```python
t = (1,2,3,4,1,3,3,5,1)

#get the frequency of particular elemnt
t.count(1)
```

Tuple Index

```python
t = (1,2,3,4,1,3,3,5,1)
print(t.index(3)) #return index of the first element is equal to 3
```

Tuple Membership

```python
#test if an item exists in a tuple or not, using the keyword in
t = (1,2,3,4,1,3,3,5,1)
print(1 in t)
print(7 in t)
```

Tuple Length

```python
t = (1,2,3,4,1,3,3,5,1)
print(len(t))
```

Tuple Sort

```python
t = (1,2,3,4,1,3,3,5,1)
new_t = sorted(t)
print(new_t)  # take elements in the tuple and return a new sorted list 
#does not sort the tuple itself
```

```python
#get the largest element in a tuple
t = (1,2,3,4,1,3,3,5,1)
print(max(t))

#get the largest element in a tuple
t = (1,2,3,4,1,3,3,5,1)
print(min(t))

#get sum of elements in the tuple
print(sum(t))
```
