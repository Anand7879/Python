Dict Creation

```python
#empty dictionary
my_dict = {}

#dictionary with integer keys
my_dict = {1: 'Anand',2: 'Shivansh'}
print(my_dict)

#dictionary with mixed keys
my_dict = {'name':'satish',1: ['abc','xyz']}
print(my_dict)

#create empty dictionary using dict()
my_dict = dict()

my_dict = dict([(1,'abc'),(2,'xyz')]) #create a dict with list of tuples
print(my_dict)
```

Dict Access

```python
my_dict = {'name':'satish','age':27,'address':'MP'}

#get name 
print(my_dict['name'])
print(my_dict['age'])
print(my_dict['address'])

```

```python
print(my_dict['degree'])
```

```python
#another way of accessing key
print(my_dict.get('address'))
```

```python
#if key is not present it will give None using get method
print(my_dict.get('degree'))
```

Dict Add or Modify Elements

```python
my_dict = {'name':'satish','age':27,'address':'MP'}

#update name
my_dict['name'] = 'Ankit'
print(my_dict)
```

```python
#add new key
my_dict['degree'] = 'B Tech'
print(my_dict)
```

Dict Delete or Remove Element

```python
my_dict = {'name':'satish','age':27,'address':'MP'}

#remove a particular item
print(my_dict.pop('age'))
print(my_dict)
```

```python
my_dict = {'name':'satish','age':27,'address':'MP'}

#remove an arbitarty key
my_dict.popitem()
print(my_dict)
```

```python
squares = {2:4,3:9,4:16,5:25}

#delete particular key
del squares[2]
print(squares)
```

```python
#remove all items
squares.clear()
print(squares)
```

```python
squares = {2:4,3:9,4:16,5:25}

#delete dictinary itself
del squares
print(squares) #NameError because dict is deleted
```

Dictionary Methods

```python
squares = {2:4,3:9,4:16,5:25}

my_dict = squares.copy()
print(my_dict)
```

```python
#fromkeys[seq[,v]] -> Return a new dictionary with keys from seq and value

subjects = {}.fromkeys(['Math','English','Hindi'],0)
print(subjects)
```

```python
subjects = {2:4, 3:9, 4:16, 5:25}
print(subjects.items()) #return a new view of the dict items(key,value)
```

```python
print(subjects.keys())
print(subjects.values())
```

Dict Comprehension

```python
d= {'a':1,'b':10,'c':12,'d':14}
for pair in d.items():
    print(pair)
```

```python
#creating a new dictionary with only pairs the value is larger than 2
d= {'a':1,'b':10,'c':12,'d':14}
new_dict = {k:v for k,v in d.items() if v>2}
print(new_dict)
```

```python
#we can also perform operations on te key value pairs
d= {'a':1,'b':10,'c':12,'d':14}
d = {k+'c':v*2 for k,v in d.items() if v>2}
print(d)
```
